# Income Statement 

```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Fetch revenue data
symbol = 'EL'
EsteeLauder = yf.Ticker(symbol)

# Get income statement data (focus on revenues)
income_statement = EsteeLauder.financials
revenue_data = income_statement.loc['Total Revenue'].transpose()

# Display the revenue data
print("Estee Lauder's Revenue Data (in billions):")
print(revenue_data)

# Check for NaN values in revenue data
if revenue_data.isnull().any():
    print("\nWarning: NaN values detected in revenue data. Handling them...")
    revenue_data = revenue_data.dropna()  # Drop rows with NaN values
    # Alternatively, you could fill NaN values with the mean or median:
    # revenue_data = revenue_data.fillna(revenue_data.mean())

# Prepare data for regression
# Convert index to datetime for better handling
revenue_data.index = pd.to_datetime(revenue_data.index)
revenue_data = revenue_data.sort_index()

# Create a numerical index for regression
X = np.array(range(len(revenue_data))).reshape(-1, 1)  # Independent variable (time)
y = revenue_data.values  # Dependent variable (revenue)

# Fit a linear regression model
model = LinearRegression()
model.fit(X, y)

# Predict future revenues (for the next 3 years)
future_X = np.array(range(len(revenue_data), len(revenue_data) + 3)).reshape(-1, 1)  # Next 3 years
predicted_revenue = model.predict(future_X)

# Combine historical and predicted revenues
future_dates = pd.date_range(start=revenue_data.index[-1] + pd.DateOffset(years=1), periods=3, freq='Y')
predicted_revenue_series = pd.Series(predicted_revenue, index=future_dates)

# Plot historical and predicted revenues
plt.figure(figsize=(10, 5))
plt.plot(revenue_data.index, revenue_data.values, label='Historical Revenue', marker='o')
plt.plot(predicted_revenue_series.index, predicted_revenue_series.values, label='Predicted Revenue', marker='o', linestyle='--')
plt.title("EsteeLauder's Revenue Prediction")
plt.xlabel("Year")
plt.ylabel("Revenue (in billions)")
plt.legend()
plt.grid()
plt.show()

# Print predicted revenues
print("\nPredicted Revenues for the Next 3 Years:")
for year, revenue in zip(predicted_revenue_series.index.year, predicted_revenue_series.values):
    print(f"{year}: ${revenue:.2f} billion")
```

# Financial Ratio 

```python
# Look at Financial Ratios

import yfinance as yf
import matplotlib.pyplot as plt
import pandas as pd  # Import pandas

# Define the ticker symbol for Apple Inc.
ticker_symbol = 'EL'

# Fetch the data
EsteeLauder = yf.Ticker(ticker_symbol)

# Get the balance sheet and income statement
balance_sheet = EsteeLauder.balance_sheet.transpose()
income_statement = EsteeLauder.financials.transpose()

# Convert the index to datetime for easier plotting
balance_sheet.index = pd.to_datetime(balance_sheet.index)
income_statement.index = pd.to_datetime(income_statement.index)

# Select important items for calculations
total_assets = balance_sheet['Total Assets']
total_liabilities = balance_sheet['Total Liabilities Net Minority Interest']
shareholder_equity = total_assets - total_liabilities  # Calculate Shareholder's Equity

net_income = income_statement['Net Income']
ordinary_shares = balance_sheet['Ordinary Shares Number']

# Calculate metrics
leverage_ratio = total_liabilities / total_assets
roe = net_income / shareholder_equity
eps = net_income / ordinary_shares

# Create a DataFrame for metrics
metrics = pd.DataFrame({
    'Leverage Ratio': leverage_ratio,
    'Return on Equity (ROE)': roe,
    'Earnings Per Share (EPS)': eps
})

# Plotting Leverage Ratio
plt.figure(figsize=(12, 6))
plt.plot(metrics.index, metrics['Leverage Ratio'], marker='o', color='blue')
plt.title(f'Leverage Ratio Over Time for {ticker_symbol}')
plt.xlabel('Date')
plt.ylabel('Leverage Ratio')
plt.xticks(rotation=45)
plt.grid()
plt.tight_layout()
plt.show()

# Plotting Return on Equity (ROE)
plt.figure(figsize=(12, 6))
plt.plot(metrics.index, metrics['Return on Equity (ROE)'], marker='o', color='green')
plt.title(f'Return on Equity (ROE) Over Time for {ticker_symbol}')
plt.xlabel('Date')
plt.ylabel('Return on Equity (ROE)')
plt.xticks(rotation=45)
plt.grid()
plt.tight_layout()
plt.show()

# Plotting Earnings Per Share (EPS)
plt.figure(figsize=(12, 6))
plt.plot(metrics.index, metrics['Earnings Per Share (EPS)'], marker='o', color='orange')
plt.title(f'Earnings Per Share (EPS) Over Time for {ticker_symbol}')
plt.xlabel('Date')
plt.ylabel('Earnings Per Share (EPS)')
plt.xticks(rotation=45)
plt.grid()
plt.tight_layout()
plt.show()
```

#  Extended Income Statement Analysis for Este Lauder (Ticker: EL)

In this section, we will:
- Import Este Lauder income statement data from Yahoo Finance
- Analyze and predict:
  - Total Revenue
  - Gross Profit
  - Net Income
- Forecast for the next 3 years
- Visualize the trends
- Summarize findings

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np

# Load financial data for Este Lauder (EL)
ticker = yf.Ticker("EL")
income_stmt = ticker.financials.T

# Select relevant columns
data = income_stmt[['Total Revenue', 'Gross Profit', 'Net Income']]
data.index = pd.to_datetime(data.index)

# Sort and preview
data = data.sort_index()
data = data / 1e9  # convert to billions
data.tail()
```

```python
# Prepare data for prediction
def prepare_prediction(df, col):
    df = df.dropna()
    df = df.reset_index()
    df['Year'] = df['index'].dt.year
    X = df[['Year']]
    y = df[col]
    return X, y, df

# Forecast next 3 years using Linear Regression
def predict_next_years(X, y, future_years=3):
    model = LinearRegression()
    model.fit(X, y)
    future_X = pd.DataFrame({'Year': np.arange(X['Year'].max()+1, X['Year'].max()+1+future_years)})
    y_pred = model.predict(future_X)
    return future_X, y_pred, model
```

```python
# Predict and plot function
def plot_predictions(df, col, future_X, y_pred):
    plt.figure(figsize=(8,5))
    plt.plot(df['Year'], df[col], marker='o', label='Actual')
    plt.plot(future_X['Year'], y_pred, marker='x', linestyle='--', label='Predicted')
    plt.title(f"{col} Prediction")
    plt.xlabel("Year")
    plt.ylabel("USD (Billions)")
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
# Revenue Prediction
X_rev, y_rev, df_rev = prepare_prediction(data, 'Total Revenue')
future_X_rev, y_pred_rev, model_rev = predict_next_years(X_rev, y_rev)
plot_predictions(df_rev, 'Total Revenue', future_X_rev, y_pred_rev)
```

```python
# Gross Profit Prediction
X_gp, y_gp, df_gp = prepare_prediction(data, 'Gross Profit')
future_X_gp, y_pred_gp, model_gp = predict_next_years(X_gp, y_gp)
plot_predictions(df_gp, 'Gross Profit', future_X_gp, y_pred_gp)
```

```python
# Net Income Prediction
X_ni, y_ni, df_ni = prepare_prediction(data, 'Net Income')
future_X_ni, y_pred_ni, model_ni = predict_next_years(X_ni, y_ni)
plot_predictions(df_ni, 'Net Income', future_X_ni, y_pred_ni)
```

###  Summary of Findings

- **Revenue** has shown volatility with a small recent decline; future growth appears moderate.
- **Gross Profit** trends similarly to revenue, indicating steady cost structures.
- **Net Income** is more volatile, reflecting factors like operating expenses or one-time charges.
- Based on linear regression, **we expect modest recovery and growth** over the next three years for all three metrics, assuming no major macroeconomic disruptions.

>  Note: These forecasts are simplistic and do not account for seasonality, external economic shocks, or Este Lauder-specific strategic changes.
